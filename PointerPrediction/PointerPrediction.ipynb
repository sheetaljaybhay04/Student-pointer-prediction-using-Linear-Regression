```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python

def plotPointers(sem, ptr):
    plt.plot(list(range(1, sem)), ptr)
    plt.xlabel('Semester')
    plt.ylabel('Pointer')
    plt.title('Your performance till now')
    plt.xticks(list(range(1, sem)))
    plt.show()
    

def main():
    sem=int(input("Enter semester for which you want to predict pointer "))
    print("Enter pointers of previous "+ str(sem-1)+" semesters:")
    ptr, ptrs, data, y, X=[], [], [], [], []
    min, max=4.00, 10.00
    
    for i in range(1, sem):
        ptr.append(float(input("Sem "+str(i)+" :")))
    ptrs.append(ptr)
    
    f=open("previousBatchData.txt")
    
    for entry in f.readlines():
        subdata=entry.strip().split(",")
        subdata=list(map(float, subdata))
        data.append(subdata)
        X.append(subdata[:sem-1])
        y.append(subdata[sem-1])
        
    X=np.array(X)
    y=np.array(y)
    
    #X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
    
    model = LinearRegression()
    model.fit(X, y)
    
    #y_pred = model.predict(X_test)
    
    ynew = model.predict(ptrs)
    ynew=sorted([min, ynew, max])[1]
    
    #err=np.sqrt(metrics.mean_squared_error(y_test, y_pred))
    #accuracy = metrics.r2_score(y_test, y_pred)
    
    plotPointers(sem, ptr)
    print("\nPredicted pointer of semester "+str(sem)+" is %.2f" % ynew[0])
    
```

```python
main()
```
